```python
autor_col = 'author.uniqueId'  
metricas = ['statsV2.diggCount', 'statsV2.commentCount', 'statsV2.shareCount', 'statsV2.playCount']

df[metricas] = df[metricas].apply(pd.to_numeric, errors ='coerce')
```

```python
agr_autor =  df.groupby(autor_col)[metricas].agg(['mean', 'median']).round(2)

renomeadas = {
    'statsV2.diggCount_mean': 'likes_media',
    'statsV2.diggCount_median': 'likes_mediana',
    'statsV2.commentCount_mean': 'comentarios_media',
    'statsV2.commentCount_median': 'comentarios_mediana',
    'statsV2.shareCount_mean': 'compart_media',
    'statsV2.shareCount_median': 'compart_mediana',
    'statsV2.playCount_mean': 'plays_media',
    'statsV2.playCount_median': 'plays_mediana',
}

agr_autor.columns = ['_'.join(col) for col in agr_autor.columns]
agr_autor = agr_autor.rename(columns=renomeadas)



display(agr_autor.head(10))
```

```python
autor = 'author.uniqueId'
metrica = 'statsV2.playCount'
min_posts = 3  # autor com pelo menos 3 posts


agg = df.groupby(autor)[metrica].agg(['mean','median','count'])
agg_filtrado = agg[agg['count'] >= min_posts]


top = agg_filtrado.sort_values('mean', ascending=False).head(8)


ax = top[['mean','median']].plot(kind='bar', figsize=(12,6),
                                 title=f'M√©dia vs Mediana de plays (Top 8, m√≠nimo {min_posts} posts)')
plt.ylabel('Plays')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Em primeiro lugar, tentei fazer5 a compara√ß√£o de m√©dia e mediana entre todos. No entanto, a maioria dos autores tinha apenas um posto, fazendo com que as duas fossem na maioria das vezes a mesma coisa. 

Como se percebe, a taylor swift possui uma consist√™ncia de plays em seus posts, sempre muito alta. O segundo autor, por outro lado, possui um outlier em que a quantidade foi significativamente maior do que o normal. 

```python
autor = 'author.uniqueId'
likes_col = 'statsV2.diggCount'
comments_col = 'statsV2.commentCount'


df[[likes_col, comments_col]] = df[[likes_col, comments_col]].apply(pd.to_numeric, errors='coerce')

#se tiver nenhum like, √© classificado como NaN
df['prop_comments_por_like'] = np.where(
    df[likes_col] > 0, 
    df[comments_col] / df[likes_col], 
    np.nan
)


resultado = (
    df.groupby(autor)['prop_comments_por_like']
      .mean()
      .sort_values(ascending=False)
      .to_frame(name='media_prop_comments_por_like')
)


contagem_validos = df.groupby(autor)['prop_comments_por_like'].count()
resultado = resultado.join(contagem_validos.rename('qtd_posts_validos'))


display(resultado.head(15))
```

```python
autor = 'author.uniqueId'
likes_col = 'statsV2.diggCount'
comments_col = 'statsV2.commentCount'
seguidores_col = 'authorStatsV2.followerCount'  


df[[likes_col, comments_col, seguidores_col]] = df[[likes_col, comments_col, seguidores_col]].apply(pd.to_numeric, errors='coerce')


df['prop_comments_por_like'] = np.where(df[likes_col] > 0, df[comments_col] / df[likes_col], np.nan)


resultado = (
    df.groupby(autor)['prop_comments_por_like']
      .mean()
      .to_frame(name='media_prop_comments_por_like')
)


resultado['qtd_posts_validos'] = df.groupby(autor)['prop_comments_por_like'].count()


seguidores_por_autor = df.groupby(autor)[seguidores_col].max().rename('seguidores')
resultado = resultado.join(seguidores_por_autor)


resultado = resultado.sort_values('media_prop_comments_por_like', ascending=False)

display(resultado.head(15))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.scatterplot(
    data=resultado,
    x='seguidores',
    y='media_prop_comments_por_like',
    hue='qtd_posts_validos',
    palette='viridis',
    alpha=0.8
)
plt.title('Seguidores vs Coment√°rios por Like (m√©dia por autor)')
plt.xlabel('Seguidores')
plt.ylabel('Coment√°rios por like (m√©dia)')
plt.legend(title='Posts v√°lidos')
plt.tight_layout()
plt.show()
```

Pode-se notar que muitos dos perfis com menos seguidores possuem uma facilidade maior de conseguir mais engajamento, visto que as pessoas realmente se sentem parte daquela comunidade e falam entre si, enquanto perfis maiores, como Taylor Swift, por exemplo, por terem muitos seguidores, encontram maior dificuldade de engajamento. Isso n√£o significa, no entanto, que ter um perfiil pequeno n√£o significa ter engajamento alto.
